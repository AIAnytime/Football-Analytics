```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
#import data
data = pd.read_csv("epl.csv")
```

```python
data.info()
```

```python
data = data[['Player', 'Pos', 'Squad', '90s', 'Prog']]
```

```python
data.head(10)
```

```python
data.info()
```

```python
data['ProgPer90'] = data['Prog']/data['90s']
```

```python
data.head(10)
```

```python
#Data Manipulation to eliminate players (Remove who has played less than 6.5 of "90 minutes" football)
data = data[data['90s']>=6.5]
```

```python
#drop GK
data = data[data['Pos'] != 'GK'].reset_index()
```

```python
#set default colors
text_color = 'white'
background = '#313332'
```

```python
fig, ax = plt.subplots(figsize=(10,5))
fig.set_facecolor(background)
ax.patch.set_facecolor(background)

#set up our base layer
mpl.rcParams['xtick.color'] = text_color
mpl.rcParams['ytick.color'] = text_color

ax.grid(ls='dotted',lw=.5,color='lightgrey',axis='y',zorder=1)
spines = ['top','bottom','left','right']
for x in spines:
    if x in spines:
        ax.spines[x].set_visible(False)

sns.swarmplot(x='ProgPer90',data=data,color='white',zorder=1)

#plot Shaw
plt.scatter(x=5.714286,y=0,c='red',edgecolor='white',s=250,zorder=2)
plt.text(s='Shaw',x=5.714286,y=.08,c='red', fontsize=12)

#plot Robertson
plt.scatter(x=7.447699,y=0,c='pink',edgecolor='white',s=250,zorder=2)
plt.text(s='Robertson',x=7.447699,y=.09,c='pink', fontsize=12)

#plot TAA
plt.scatter(x=8.682927,y=0,c='yellow',edgecolor='white',s=250,zorder=2)
plt.text(s='TAA',x=8.682927,y=-.04,c='yellow', fontsize=12)

#plot Tierney
plt.scatter(x=4.497041,y=0,c='green',edgecolor='white',s=250,zorder=2)
plt.text(s='Tierney',x=4.497041,y=-.07,c='green', fontsize=12)

#plot Cancelo
plt.scatter(x=7.04142,y=0,c='blue',edgecolor='white',s=250,zorder=2)
plt.text(s='Cancelo',x=7.04142,y=-.07,c='blue', fontsize=13)

#plot Walker
plt.scatter(x=4.661654,y=0,c='purple',edgecolor='white',s=250,zorder=2)
plt.text(s='Walker',x=4.661654,y=.12,c='purple', fontsize=13)

#plot AWB
plt.scatter(x= 3.727273,y=0,c='magenta',edgecolor='white',s=250,zorder=2)
plt.text(s='AWB',x= 3.727273,y=-.09,c='magenta', fontsize=12)

#plot Reece
plt.scatter(x=4.87013,y=0,c='cyan',edgecolor='white',s=250,zorder=2)
plt.text(s='Reece',x=4.87013,y=.07,c='cyan', fontsize=12)

plt.title('Progressive Passes by the top Full backs in the Premier League 2020/21',c='white',fontsize=15,
          bbox={'facecolor': 'red', 'alpha': 0.6, 'pad': 10})

plt.xlabel('Progressive Passes per 90 minutes',c='yellow', fontsize=14)

fig.text(.12,.03,"@Sonu", fontstyle='italic',fontsize=12, fontfamily='Andale Mono',color=text_color)

plt.savefig('output.png')
```
